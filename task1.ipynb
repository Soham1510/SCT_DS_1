```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('API_SP.POP.TOTL.csv')
df.info()
```

```python
df.drop(['Indicator Name' , 'Indicator Code' , 'Country Code'], axis=1,inplace=True)
```

```python
df
```

```python
#India Data 

india_data = df.loc[df['Country Name']=="India"].T
india_data = india_data.reset_index()
india_data.columns = ['Year', 'Population']
india_data = india_data[1:]
india_data['Year'] = india_data['Year'].astype(int)
india_data['Population'] = india_data['Population'].astype(float)

india_data.isna().sum()
india_data
```

```python
# 📊 Bar Chart - Population Over Time

plt.figure(figsize=(12,6))
plt.bar(india_data['Year'], india_data['Population']/1e9, color="Pink")
plt.title("India's Population Over Time", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Population (Billions)")
plt.show()
```

```python
# 📊 Histogram - Distribution of Population Values
plt.figure(figsize=(10,6))
sns.histplot(india_data['Population']/1e9, bins=10, color="red", edgecolor="red", kde=True)
plt.title("Distribution of India's Population", fontsize=16)
plt.xlabel("Population (Billions)")
plt.ylabel("Frequency")
plt.show()
```

```python
grp = df.groupby('Country Name')
country_list=[]
for group, data in grp:
    # print(group)
    country_list.append(group)
    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def show():
    for i in country_list:
        data = df.loc[df['Country Name']==i].T
        data = data.reset_index()
        data.columns = ['Year', 'Population']
        data = data[1:]
        data['Year'] = data['Year'].astype(int)
        data['Population'] = data['Population'].astype(float)
        
        # Clean NaN
        data = data.dropna()

        # 📊 Bar Chart
        print("📊 Bar Chart - Population Over Time")
        plt.figure(figsize=(12,6))
        plt.bar(data['Year'], data['Population']/1e9, color="skyblue")
        plt.title(f"{i} Population Over Time", fontsize=16)
        plt.xlabel("Year")
        plt.ylabel("Population (Billions)")
        plt.show()

        print('\n\n')

        # 📊 Histogram with KDE
        print('📊 Histogram - Distribution of Population Values')
        plt.figure(figsize=(10,6))
        sns.histplot(data['Population']/1e9, bins=10, color="orange", edgecolor="black", kde=True)
        plt.title(f"Distribution of {i} Population", fontsize=16)
        plt.xlabel("Population (Billions)")
        plt.ylabel("Frequency")
        plt.show()

show()
```
